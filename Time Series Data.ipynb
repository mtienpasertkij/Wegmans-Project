```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
mydir = "/Users/ptienpasertkij/Documents/RIT/Summer 2021/BANA785/All Data Files/"
wegmansData = mydir + "WegmansFinalCleaned.txt"
wegmans = pd.read_csv(wegmansData, sep="\t")
```

```python
wegmans = wegmans.drop(columns=['TRANSACTION_KEY', 'transaction'])
```

```python
wegmans.dtypes
```

```python
wegmans.loc[wegmans['HH_INCOME'] == '<10K', 'HH_INCOME'] = 10000
wegmans.loc[wegmans['HH_INCOME'] == '15K-20K', 'HH_INCOME'] = 15000
wegmans.loc[wegmans['HH_INCOME'] == '20K-30K', 'HH_INCOME'] = 20000
wegmans.loc[wegmans['HH_INCOME'] == '30K-40K', 'HH_INCOME'] = 30000
wegmans.loc[wegmans['HH_INCOME'] == '40K-50K', 'HH_INCOME'] = 40000
wegmans.loc[wegmans['HH_INCOME'] == '50K-75K', 'HH_INCOME'] = 50000
wegmans.loc[wegmans['HH_INCOME'] == '75K-100K', 'HH_INCOME'] = 75000
wegmans.loc[wegmans['HH_INCOME'] == '100K-125K', 'HH_INCOME'] = 100000
wegmans.loc[wegmans['HH_INCOME'] == '125K+', 'HH_INCOME'] = 125000
wegmans['HH_INCOME'].unique()
```

```python
def isorganic(wegmans):
  if wegmans['IS_ORGANIC'] & wegmans['IS_WEGMANSBRAND'] == 1:
    return 1
  elif wegmans['IS_ORGANIC'] == 0:
    return 0
  elif wegmans['IS_WEGMANSBRAND'] == 0:
    return 0
  else:
    return 'Undefined'

wegmans['WEGMANS_ORGANIC'] = wegmans.apply(isorganic, axis=1)
```

```python
def isfampack(wegmans):
  if wegmans['IS_FAMILYPACK'] & wegmans['IS_WEGMANSBRAND'] == 1:
    return 1
  elif wegmans['IS_FAMILYPACK'] == 0:
    return 0
  elif wegmans['IS_WEGMANSBRAND'] == 0:
    return 0
  else:
    return 'Undefined'

wegmans['WEGMANS_FAMPACK'] = wegmans.apply(isfampack, axis=1)
```

```python
def isbulk(wegmans):
  if wegmans['UNITS'] >= 6:
    return 1
  else:
    return 0

wegmans['IS_BULK'] = wegmans.apply(isbulk, axis=1)
```
