```python
wegmans = wegmans.drop(columns=['TRANSACTION_KEY', 'transaction'])
```

```python
wegmanssort = wegmans
```

```python
wegmanssort
```

```python
wegmanssort.DATE_KEY = pd.to_datetime(wegmans.DATE_KEY,format='%Y-%m-%d')
```

```python
wegmanssort = wegmanssort.sort_values(['DATE_KEY','CATEGORY_NAME']) 
```

```python
wegmansnew = wegmanssort.groupby([pd.Grouper(key='DATE_KEY',freq='d'), 'CATEGORY_NAME'])['SALES'].sum().reset_index()
```

```python
wegmansnew['Yesterday Sales'] = wegmansnew.groupby(['CATEGORY_NAME'])['SALES'].shift()
wegmansnew['Week Ago Sales'] = wegmansnew.groupby(['CATEGORY_NAME'])['SALES'].shift(7)
```

```python
wegmansnew['isWeekend'] = (wegmansnew['DATE_KEY'].dt.dayofweek >= 5)
```

```python
wegmansnew.isna().sum()
```

```python
wegmansnew['Yesterday Sales'].fillna(0, inplace = True)
wegmansnew['Week Ago Sales'].fillna(0, inplace = True)
```

```python
wegmansnew
```

```python
wegmansnew.groupby('CATEGORY_NAME')['SALES'].sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
```

```python
from sklearn.metrics import mean_squared_error
def rmsle(ytrue, ypred):
    return np.sqrt(mean_squared_error(ytrue, ypred))
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(wegmansnew['CATEGORY_NAME'])
wegmansnew['CATCODES'] = le.transform(wegmansnew['CATEGORY_NAME'])
```

```python
import datetime
pd.set_option('mode.chained_assignment',None) #Turn off SettingWithCopyWarning

start_date = datetime.date(2019,1,7)
end_date = datetime.date(2020,3,1)
delta = datetime.timedelta(days=1)
windowdelta = datetime.timedelta(days=30)
dateidx = start_date

mean_error = []
wegmanspred = pd.DataFrame() 

while dateidx < end_date:
    print(dateidx)
    #we will now predict with a 30 day prior window
    train = wegmansnew[(wegmansnew['DATE_KEY'] < pd.Timestamp(dateidx)) & (wegmansnew['DATE_KEY'] > pd.Timestamp(dateidx - windowdelta))]
    val = wegmansnew[wegmansnew['DATE_KEY'] == pd.Timestamp(dateidx)] #predictions for 2019
    
    if val.size > 0:
        xtr, xts = train[['CATCODES','Yesterday Sales','Week Ago Sales','isWeekend']], val[['CATCODES','Yesterday Sales','Week Ago Sales','isWeekend']]
        ytr, yts = train['SALES'].values, val['SALES'].values
    
        mdl = RandomForestRegressor(n_estimators=500, n_jobs=-1, random_state=0)
        mdl.fit(xtr, ytr)

        p = mdl.predict(xts)
    
        val.loc[:, 'Prediction'] = np.round(p)
        wegmanspred = pd.concat([wegmanspred,val])
        error = rmsle(yts, p)
        print('Day %s - Error %.5f' % (str(dateidx), error))
        mean_error.append(error)
    dateidx += delta
print('Mean Error = %.5f' % np.mean(mean_error))
```

```python
wegmanspred.to_csv('prepandemic_30daywindow.csv')
```

```python
# Build predictions for 2019, then compare with actual for 2020
import datetime
pd.set_option('mode.chained_assignment',None)

start_date = datetime.date(2019,1,7)
end_date = datetime.date(2020,3,1)
delta = datetime.timedelta(days=1)
windowdelta = datetime.timedelta(days=30)
dateidx = start_date

mean_error = []
wegmanspredlgbm = pd.DataFrame() 

while dateidx < end_date:
    print(dateidx)
    train = wegmansnew[(wegmansnew['DATE_KEY'] < pd.Timestamp(dateidx)) & (wegmansnew['DATE_KEY'] > pd.Timestamp(dateidx - windowdelta))]
    val = wegmansnew[wegmansnew['DATE_KEY'] == pd.Timestamp(dateidx)]
    
    if val.size > 0:
        xtr, xts = train[['CATCODES','Yesterday Sales','Week Ago Sales','isWeekend']], val[['CATCODES','Yesterday Sales','Week Ago Sales','isWeekend']]
        ytr, yts = train['SALES'].values, val['SALES'].values
    
        mdl = LGBMRegressor(n_estimators=1000, learning_rate=0.01)
        mdl.fit(xtr, np.log1p(ytr))

        p = np.expm1(mdl.predict(xts))
    
        val.loc[:, 'Prediction'] = np.round(p)
        wegmanspredlgbm = pd.concat([wegmanspredlgbm,val])
        error = rmsle(yts, p)
        print('Day %s - Error %.5f' % (str(dateidx), error))
        mean_error.append(error)
    dateidx += delta
print('Mean Error = %.5f' % np.mean(mean_error))
```

```python
# Migrate 2019 predictions to 2020
import datetime
pd.set_option('mode.chained_assignment',None) #Turn off SettingWithCopyWarning

start_date = datetime.date(2020,3,1)
end_date = datetime.date(2020,12,31)
delta = datetime.timedelta(days=1)
dateidx = start_date

#just predict all days using final xts model from above
while dateidx <= end_date:
    print(dateidx)
    val = wegmansnew[wegmansnew['DATE_KEY'] == pd.Timestamp(dateidx)] 
    if val.size > 0:
        xts = val[['CATCODES','Yesterday Sales','Week Ago Sales','isWeekend']]
        p = mdl.predict(xts)
        val.loc[:, 'Prediction'] = np.round(p)
        wegmanspredlgbm = pd.concat([wegmanspredlgbm,val])
    dateidx += delta
```
