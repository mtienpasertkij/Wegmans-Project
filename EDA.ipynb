```python
x_children = customer['HH_CHILDREN'].value_counts().index
y_children = customer['HH_CHILDREN'].value_counts().values

plt.bar(x_children,y_children)

def addlabels(x,y):
    for i in range(len(x)):
        plt.text((i-.25),y[i],y[i])
        
addlabels(x_children, y_children)

plt.title("Number of Children in the Household")

plt.xlabel("Number of Children")
plt.ylabel("Number of Customers")

plt.show()
```

```python
x_age = customer['HOH_AGE'].value_counts().index
y_age = customer['HOH_AGE'].value_counts().values

plt.bar(x_age,y_age)

plt.title("Head of Household Age")

plt.xlabel("Age")
plt.ylabel("Number of Customers")

plt.show()
```

```python
x_income = customer['HH_INCOME'].value_counts().index
y_income = customer['HH_INCOME'].value_counts().values

plt.bar(x_income,y_income,10000)

plt.title("Household Income")

plt.xlabel("Income")
plt.ylabel("Number of Customers")

plt.show()
```

```python
sales["DATE_KEY"] =  pd.to_datetime(sales["DATE_KEY"])
sales["DATE_KEY"].dt.year.value_counts(sort = False).plot.bar()

plt.title("Number of Items by Year")

plt.xlabel("Year")
plt.ylabel("Number of Items")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_ylabel('Population(in million)')
ax.set_title('Years')
ax.set_xticks(pd.to_datetime(sales["DATE_KEY"]))
ax.set_xticklabels(sales["DATE_KEY"].dt.year.value_counts(sort = False))
```
