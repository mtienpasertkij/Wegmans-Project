```python
WFD1 = WFD.groupby([pd.Grouper(key='DATE_KEY',freq='d'),'CATEGORY_NAME'])['SALES'].sum().reset_index()
```

```python
WFD1['Yesterday Sales_cat'] = WFD.groupby(['CATEGORY_NAME'])['SALES'].shift()
WFD1['Week Ago Sales_cat'] = WFD.groupby(['CATEGORY_NAME'])['SALES'].shift(7)
WFD1['Yesterday Sales_deptcat'] = WFD.groupby(['DEPT_CAT'])['SALES'].shift()
WFD1['Week Ago Sales_deptcat'] = WFD.groupby(['DEPT_CAT'])['SALES'].shift(7)
WFD1['quarter'] = WFD['DATE_KEY'].dt.quarter
WFD1['weekno'] = WFD['DATE_KEY'].dt.week
WFD1['dayofweek'] = WFD['DATE_KEY'].dt.weekday
WFD1['year'] = WFD['DATE_KEY'].dt.year
```

```python
from sklearn.metrics import mean_squared_error
def rmsle(ytrue, ypred):
    return np.sqrt(mean_squared_error(ytrue, ypred))
```

```python
WFD1.groupby('CATEGORY_NAME')['SALES'].sum()
```

```python
import datetime
start_date = datetime.date(2019,1,14)
end_date = datetime.date(2019,10,25)
delta = datetime.timedelta(days=1)
windowdelta = datetime.timedelta(days=30)
dateidx = start_date

mean_error = []

while dateidx < end_date:
    #print(dateidx)
    train = WFD1[(WFD1['DATE_KEY'] < pd.Timestamp(dateidx)) & (WFD1['DATE_KEY'] > pd.Timestamp(dateidx - windowdelta))]
    val = WFD1[WFD1['DATE_KEY'] == pd.Timestamp(dateidx)]
    
    p = val['Yesterday Sales_cat'].values

    error = rmsle(val['SALES'].values, p)
 
    print('Day %s - Error %.5f' % (str(dateidx), error))
    mean_error.append(error)
    dateidx += delta
print('Mean Error = %.5f' % np.mean(mean_error))
```
