- Examine the data types
- Look at unique values in HH_INCOME and changing the format of the data

    - EX: <10K -> 10000; 15K-20K -> 15000; 20K-30K -> 20000, so on

        - use mean for HH_INCOME
            - Chose mean after plotting a boxplot - there are no outliers in the dataset
        - use median for HH_CHILDREN
            - Chose median so outliers can't affect our future analysis
- Export datasets into CSV
    - Wegmans Customer Data without NA.csv
    - Wegmans Customer Data with NA.csv

test = test.groupby('TRANSACTION_KEY')['Sales'].sum().reset_index()

```python
test.isna().sum()
```

```python
test = test[(test['SALES'] >= 0) | (test['SALES'].isnull())]
```

```python
test.dtypes
```

```python
# to explicitly convert the date column to type DATETIME
test['DATE_KEY'] = pd.to_datetime(test['DATE_KEY'])
test = test.set_index('DATE_KEY')
```

```python
test.describe()
```

```python
test
```

```python
#create feature to identify whether a purchased item was wegmans organic brand
def isorganic(test):
  if test['IS_ORGANIC'] & test['IS_WEGMANSBRAND'] == 1:
    return 1
  elif test['IS_ORGANIC'] == 0:
    return 0
  elif test['IS_WEGMANSBRAND'] == 0:
    return 0
  else:
    return 'Undefined'
```

```python
#create feature to identify whether a purchased item was wegmans brand family pack
def isfampack(test):
  if test['IS_FAMILYPACK'] & test['IS_WEGMANSBRAND'] == 1:
    return 1
  elif test['IS_FAMILYPACK'] == 0:
    return 0
  elif test['IS_WEGMANSBRAND'] == 0:
    return 0
  else:
    return 'Undefined'
```

```python
test['WEGMANS_ORGANIC'] = test.apply(isorganic, axis=1)
```

```python
test['WEGMANS_FAMPACK'] = test.apply(isfampack, axis=1)
```

```python
test['WEGMANS_ORGANIC'].value_counts()
```

```python
test['UNITS'].value_counts()
```

```python
test['WEGMANS_FAMPACK'].value_counts()
```

```python
#create feature to identify whether an item was bought in bulk
def isbulk(test):
  if test['UNITS'] >= 6:
    return 1
  else:
    return 0
```

```python
test['IS_BULK'] = test.apply(isbulk, axis=1)
```

```python
test['IS_BULK'].value_counts()
```

```python
test.dtypes
```

```python
# Define which columns should be encoded vs scaled
columns_to_encode = ['UNITS','ITEM_DESCRIPTION','SALES','PRODUCT_HIERARCHY','DEPARTMENT_NAME','CATEGORY_NAME','CLASS_NAME','DEPTCAT','DEPTCLASS','CATCLASS',]
columns_to_scale  = ['UNITS','WEIGHT','SALES','PPU','HOH_AGE','HH_INCOME','HH_CHILDREN']
other_cols = ['TRANSACTION_KEY','CUSTOMER_ID','ITEM_NBR','IS_ORGANIC','IS_FAMILYPACK','IS_WEGMANSBRAND','HOUSEHOLD_ID','WEGMANS_ORGANIC','WEGMANS_FAMPACK','IS_BULK']
```

```python
# Instantiate encoder/scaler
scaler = StandardScaler()
ohe    = OneHotEncoder(sparse=False)
```

```python
# Scale and Encode Separate Columns
scaled_columns  = scaler.fit_transform(test[columns_to_scale]) 
encoded_columns = ohe.fit_transform(test[columns_to_encode])
```

```python
other_cols
```

```python
# Concatenate (Column-Bind) Processed Columns Back Together
processed_test = np.concatenate([scaled_columns, encoded_columns], axis=1)
```
