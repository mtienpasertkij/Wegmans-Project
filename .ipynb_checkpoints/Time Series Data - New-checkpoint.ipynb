```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
mydir = "/Users/ptienpasertkij/Documents/RIT/Summer 2021/BANA785/All Data Files/"
wegmansData = mydir + "Wegmans All Data.txt"
wegmans = pd.read_csv(wegmansData, sep=",")
```

```python
import datetime
```

```python
wegmans.DATE_KEY = pd.to_datetime(wegmans.DATE_KEY,format='%Y%m%d')
wegmans['MONTH'] = wegmans['DATE_KEY'].dt.month
wegmans['YEAR'] = wegmans['DATE_KEY'].dt.year
```

```python
wegmans['monthyear'] = pd.to_datetime(wegmans[['YEAR', 'MONTH']].assign(DAY=1))
```
